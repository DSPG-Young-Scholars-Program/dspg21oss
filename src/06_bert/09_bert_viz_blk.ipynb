```python
import os 
import altair as alt
import pandas as pd 

os.chdir("/sfs/qumulo/qhome/kb7hp/git/dspg21oss/data/dspg21oss/")
df = pd.read_csv("word_embedding_viz.csv")
```

```python
keywords = ['database', "machine","learning", "artificial", "intelligence","python", "r", "ai", "ml", "visual"]

x = ["database", "ai", "ml", "visual"]
df_sub = df[df.word.isin(x)]

color_by = 'word'

# Make the plot
alt.Chart(df_sub, title=f"Word Similarity: AI, ML, database, visualization").mark_circle(size=100).encode(
    alt.X('x',
        scale=alt.Scale(zero=False)
    ), y="y",
    color= color_by,
    size=alt.Size('stars'),
    #href="link",
    tooltip=['slug', 'word', 'description', 'stars']
    ).interactive().properties(
    width=500,
    height=500
)
```
